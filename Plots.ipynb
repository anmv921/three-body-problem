```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit
import os
import time
import pandas as pd
plt.style.use("seaborn-bright")
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["xtick.top"] = "true"
plt.rcParams["ytick.right"] = "true"
```

```python
df = pd.read_csv("data.dat", sep=",", header=0, index_col=None)

t = df["t"].values

x1 = df["x1"].values
y1 = df["y1"].values
z1 = df["z1"].values

x2 = df["x2"].values
y2 = df["y2"].values
z2 = df["z2"].values

x3 = df["x3"].values
y3 = df["y3"].values
z3 = df["z3"].values
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(x1, y1, z1)
ax.scatter(x2, y2, z2)
ax.scatter(x3, y3, z3)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
